## **Pré-processamento de imagens**

---








#**Introdução**#
O processamento de imagens é um ramo da visão computacional que utiliza vários algoritmos para manipular e analisar imagens digitais. Envolve o uso de operações matemáticas ou estatísticas para modificar imagens para muitas aplicações, incluindo, entre outras, imagens médicas e de satélite e fotografia digital.


# **Exemplos de bibliotecas/frameworks**
**1. OpenCV**

É usada principalmente em tarefas de visão computacional, como detecção de objetos, detecção de rosto, reconhecimento de rosto, segmentação de imagem, etc., mas também contém muitas funções úteis em ML.

**2. scikit-image**

Biblioteca de processamento de imagem de código aberto para a linguagem de programação Python. Inclui algoritmos para segmentação, transformações geométricas, manipulação de espaço de cores, análise, filtragem, morfologia, detecção de recursos e muito mais.

**3. Scipy**

Originalmente projetado para cálculos matemáticos e científicos, o SciPy também é uma biblioteca importante para realizar processamento de imagens multidimensionais, importando o submódulo scipy.ndimage. SciPy fornece funções para operar em arrays Numpy n-dimensionais.


# **Exemplo de aplicação**

 O código a seguir está exibindo os canais de cores separados (vermelho, verde e azul) de uma imagem. Cada canal de cor é exibido como uma imagem separada. Isso pode ser útil para entender como diferentes características da imagem são capturadas em cada canal de cor. Por exemplo, áreas da imagem que são muito vermelhas terão valores altos no canal vermelho e valores mais baixos nos canais verde e azul.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
img = cv.imread('goku.jpeg')
b, g, r = cv.split(img)

fig, ax = plt.subplots(1, 3, figsize=(16, 8))
fig.tight_layout()

ax[0].imshow(cv.cvtColor(r, cv.COLOR_BGR2RGB))
ax[0].set_title("Red")

ax[1].imshow(cv.cvtColor(g, cv.COLOR_BGR2RGB))
ax[1].set_title("Green")

ax[2].imshow(cv.cvtColor(b, cv.COLOR_BGR2RGB))
ax[2].set_title("Blue")
```

## **Segmentação de Imagens**

---



# **Introdução**
A segmentação de imagens é uma técnica comumente usada no processamento e análise de imagens digitais para particionar uma imagem em múltiplas partes ou regiões, geralmente com base nas características dos pixels da imagem. A segmentação de imagens pode envolver a separação do primeiro plano do fundo ou o agrupamento de regiões de pixels com base em semelhanças de cor ou forma. Por exemplo, uma aplicação comum de segmentação de imagens em imagens médicas é detectar e rotular pixels em uma imagem ou voxels de um volume 3D que represente um tumor no cérebro ou em outros órgãos de um paciente .

# **Exemplos de bibliotecas/frameworks**
**1. TensorFlow**

 Permite a implementação de redes neurais profundas para segmentação semântica de imagens.

**2. OpenCV**

 Oferece várias técnicas de segmentação de imagens, como segmentação baseada em cores, textura, etc.

**3. Scikit-image**

 Fornece vários algoritmos de segmentação de imagens.

# **Exemplo de aplicação**

 O código a seguir está exibindo uma imagem de amostra e sua versão convertida em HSV lado a lado para comparação. A conversão para HSV é útil em muitas aplicações de processamento de imagens, pois fornece uma representação mais intuitiva das cores na imagem em comparação com o formato RGB.

```python
from skimage import data
from skimage.color import rgb2hsv
import matplotlib.pyplot as plt


plt.figure(figsize=(15, 15))

coffee = data.coffee()
plt.subplot(1, 2, 1)

plt.imshow(coffee)

hsv_coffee = rgb2hsv(coffee)
plt.subplot(1, 2, 2)

hsv_coffee_colorbar = plt.imshow(hsv_coffee)

plt.colorbar(hsv_coffee_colorbar, fraction=0.046, pad=0.04)
```

## **Detecção/classificação de imagens**

---



# **Introdução**

A Classificação de Imagens é uma tarefa fundamental no reconhecimento visual que visa compreender e categorizar uma imagem como um todo sob um rótulo específico. Ao contrário da detecção de objetos , que envolve a classificação e localização de vários objetos em uma imagem, a classificação de imagens normalmente se refere a imagens de um único objeto.

# **Exemplos de bibliotecas/frameworks**

**1. TensorFlow**

Permite a implementação de redes neurais profundas para detecção e classificação de imagens.

**2. Keras**

 Uma API de alto nível para construção e treinamento de modelos de aprendizado profundo.

**2. PyTorch**

Biblioteca Python de aprendizado de máquina de código aberto usada para implementações de aprendizado profundo, como visão computacional (usando TorchVision) e processamento de linguagem natural.

# **Exemplo de aplicação**

O código a seguir está usando um modelo de aprendizado profundo para classificar uma imagem como sendo de um carro ou de um avião. A imagem é pré-processada para estar no formato correto esperado pelo modelo, e o resultado da predição é impresso.

```python
from keras.models import load_model
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.applications.vgg16 import preprocess_input
from keras.applications.vgg16 import decode_predictions
from keras.applications.vgg16 import VGG16
import numpy as np

from keras.models import load_model

model = load_model('model_saved.h5')

image = load_img('v_data/test/planes/5.jpg', target_size=(224, 224))
img = np.array(image)
img = img / 255.0
img = img.reshape(1,224,224,3)
label = model.predict(img)
print("Predicted Class (0 - Cars , 1- Planes): ", label[0][0])
```
