2. Utilizando o matplotlib, utilize a função subplot() para mostrar 6 gráficos.

```python
x = np.arange(-1000,1000, 1)

y1 = x ** 2
y2 = x ** 5
y3 = x ** 10
y4 = x + 2
y5 = -x ** 8
y6 = x * 2

y = [y1, y2, y3, y4, y5, y6]

figura = plt.figure(figsize=(10,10))

for i in range(6):
  figura.add_subplot(3,3, i+1)
  plt.plot(x, y[i])

plt.show()
```

4. Complete o código:

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(ncols=2, nrows=2, figsize=(5.5, 3.5),
layout="constrained")

for row in range(2):
  for col in range(2):
    axs[row, col].annotate(f'axs[{row}, {col}]', (0.5, 0.5),
    transform=axs[row, col].transAxes,
    ha='center', va='center', fontsize=18,
    color='darkgrey')
    fig.suptitle('plt.subplots()')
```

8. Utilizando pandas, como lidar com valores ausentes (NaN) em um DataFrame?


```python
#Existem diversas diversas formas de lidar com valores
#ausentes, neste caso optei em excluir a linha inteira
#que possui valores NaN, utilizando o método df.dropna().

import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/bootcamp-ml/911.csv")

#soma dos valores nulos encontrados
df.isnull().sum()

#visiualizar shape antes
shape_antes = df.shape
print(shape_antes)

df.dropna(subset=['twp'], inplace = True)

#visiualizar shape depois
shape_depois = df.shape
print(shape_depois)
```
